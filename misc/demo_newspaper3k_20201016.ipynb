# Scraping

```python
import requests
from bs4 import BeautifulSoup
import re
import polars as pl

# Function to check if a string contains the word 'drugs'
def is_drug_related(string):
    return 'drugs' in string.lower()

# Send a GET request to the website
response = requests.get('https://www.infobae.com/')

# Parse the HTML content of the webpage with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all article elements
articles = soup.find_all('article')

# Initialize lists to store the data
titles = []
links = []

# Go through each article element
for article in articles:
    # Extract the title and URL of the article
    title = article.get('title')
    url = article.get('href')
    
    # Check if the title is drug-related
    if title and is_drug_related(title):
        # Store the title and URL
        titles.append(title)
        links.append(url)

# Create a Polars DataFrame with the data
df = pl.DataFrame({
    'title': titles,
    'link': links
})
```

```python
from newspaper import Article, Config, build
from bs4 import BeautifulSoup
import polars as pl
import datetime
import os

# Function to extract text from HTML using BeautifulSoup
def extract_text(html):
    soup = BeautifulSoup(html, 'html.parser')
    paragraphs = soup.find_all('p')
    text = ' '.join([p.get_text() for p in paragraphs])
    return text

# List of keywords to check for
keywords = ['crimen organizado', 'narcotráfico', 'narcotrafico', 'cocaina', 'traficante', 'crak', 'fentanilo']

# Function to check if a string contains any of the keywords
def is_drug_related(string):
    return any(keyword in string.lower() for keyword in keywords)

# Configure newspaper
config = Config()
config.fetch_images = False
config.memoize_articles = False

# List of URLs to get articles from
urls = [
    'https://www.infobae.com/',
    'https://www.clarin.com/',
    'https://www.ellitoral.com/',
    # Add more URLs as needed
]

# Initialize lists to store the data
dates = []
contents = []
links = []

# Go through each URL in the list
for url in urls:
    try:
        # Build the newspaper
        paper = build(url, config=config)

        # Go through each article in the newspaper
        for article in paper.articles:
            # Download and parse the article
            article.download()
            article.parse()

            # Extract the text from the article's HTML
            text = extract_text(article.html)

            # Check if the article's text contains any of the keywords
            if is_drug_related(text):
                # Store the article's publication date, text, and URL
                date = article.publish_date if article.publish_date else datetime.datetime.now()
                url = article.url if article.url else url
                dates.append(date)
                contents.append(text)
                links.append(url)
                print(text)

    except Exception as e:
        print(f'An error occurred while processing {url}: {e}')

# Create a Polars DataFrame with the data
df = pl.DataFrame({
    'date': dates,
    'content': contents,
    'link': links
})

# Check if the 'data' directory exists and create it if it doesn't
if not os.path.exists('data'):
    os.makedirs('data')

# Save the DataFrame to a file
filename = os.path.join('data', 'news_' + datetime.datetime.now().strftime('%Y_%m_%d') + '.csv')
df.write_csv(filename)
print(f'Saved data to {filename}')
```
